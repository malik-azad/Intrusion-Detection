```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load the dataset and display the first few rows
df = pd.read_csv('data/Train_data.csv')
print(df.head())
```

```python

# Map categorical features to numerical values
df['protocol_type'] = df['protocol_type'].map({'tcp': 0, 'udp': 1, 'icmp': 2})
df['service'] = df['service'].astype('category').cat.codes
df['flag'] = df['flag'].astype('category').cat.codes
df['class'] = df['class'].map({'normal': 0, 'anomaly': 1})
```

```python
# Check for null values and drop if any
df.dropna(inplace=True)
```

```python

# Separate features and target
X = df.drop('class', axis=1)
y = df['class']
```

```python

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python

# Train a Random Forest model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python

# Make predictions and evaluate the model
predictions = model.predict(X_val)
print(classification_report(y_val, predictions))
print('Accuracy:', accuracy_score(y_val, predictions))
```
